```python
func, inv_func = get_optimus_fn(10, 0.5)
```

```python
def plot_func(f, start, end):
    x = np.arange(start,end,1)
    y = [f(v) for v in x]
    plt.plot(x,y)
    plt.grid()
```

```python
plot_func(func,0, 50)
```
