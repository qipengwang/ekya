```python
PERIOD=200
RES=1
USE_ORACLE_INFO = True
BASE_DIR = 'results_0803_cityscapes_fixedcache/'
styles = ['^-', 'v-', 'o-']
scheds = ['inference_only', 'fair_dumb','thief']
labels = ["No-retrain", "Fair Scheduler", "Ekya"]
```

```python
all_cities_data = []
num_cities = [1,2,3,4,5,10]
for n_cities in num_cities:
    data_path = os.path.join(BASE_DIR, 'data_eval_multicam_all_data_{}_{}_{}cities.csv'.format(dataset, pred_mode,n_cities))
    all_data = pd.read_csv(data_path)
    all_data['num_cities']=n_cities
    d = all_data
    d = d[d['use_oracle'] == USE_ORACLE_INFO]
    d = d[d['period'] == PERIOD]
    # Group data and create overall_mean col
    plot_data = d.groupby(['sched', 'res', 'city', 'num_cities']).mean()
    tasks = [x for x in plot_data.columns if "task_" in x]
    plot_data['overall_mean'] = plot_data[tasks].mean(axis=1)
    all_cities_data.append(plot_data)
all_cities_data=pd.concat(all_cities_data)
```

```python
mean_acc_data = all_cities_data['overall_mean'].mean(level=['sched', 'res', 'num_cities']).unstack(level='sched')
```
