# Multi Model

```python
plt.figure()
contention_acc = 12.5
cs = 0.5 # contention slowdown
A_post_acc = 79
A_train_time = 25
A_mean_acc = 63.96

B_post_acc = 65
B_train_time = 15
B_mean_acc = 51

total_time = 300

data_A = [(contention_acc, 0), (contention_acc, A_train_time),
          [A_post_acc*cs, A_train_time], [A_post_acc*cs, A_train_time+B_train_time],
          [A_post_acc, A_train_time+B_train_time], [A_post_acc, total_time]]
data_A_mean = [(A_mean_acc,0), (A_mean_acc,300)]

data_B = [(contention_acc, 0), (contention_acc, B_train_time + A_train_time),
          [B_post_acc, A_train_time+B_train_time], [B_post_acc, total_time]]
data_B_mean = [(B_mean_acc,0), (B_mean_acc,300)]

y_data, x_data = list(zip(*data_A))
plt.plot(x_data, y_data, 'b-', label='Model A Accuracy')
y_data, x_data = list(zip(*data_A_mean))
plt.plot(x_data, y_data, 'b--', label='Model A Mean Accuracy')

y_data, x_data = list(zip(*data_B))
plt.plot(x_data, y_data, 'r-', label='Model B Accuracy')
y_data, x_data = list(zip(*data_B_mean))
plt.plot(x_data, y_data, 'r--', label='Model B Mean Accuracy')

plt.ylim([0,100])
plt.xlabel("Time")
plt.ylabel("Accuracy")
plt.title("Multi Job Our - A allocated more resources than B")
plt.legend()
```
